```python
# ðŸ›  If needed (in Jupyter only):
# !pip install spacy textblob
# !python -m spacy download en_core_web_sm

import pandas as pd
import spacy
from textblob import TextBlob

# âœ… Load spaCy model
nlp = spacy.load("en_core_web_sm")

# âœ… Load the Amazon reviews CSV
df = pd.read_csv("Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products.csv", encoding="ISO-8859-1")

# âœ… Check actual column names (just once)
print("ðŸ§¾ CSV Columns:")
print(df.columns.tolist())  # Uncomment this to inspect if unsure

# âœ… Use the correct column that contains the reviews
# Most common: 'reviews.text'
review_col = 'reviews.text'

# Drop missing reviews and sample 5
sample_reviews = df[review_col].dropna().sample(5, random_state=1).tolist()

# âœ… Process each review
for i, review in enumerate(sample_reviews):
    print(f"\nðŸ“ Review {i+1}: {review}\n")

    # --- Named Entity Recognition ---
    doc = nlp(review)
    entities = [(ent.text, ent.label_) for ent in doc.ents if ent.label_ in ["PRODUCT", "ORG"]]

    print("ðŸ§  Named Entities (Products/Brands):")
    if entities:
        for ent in entities:
            print(f"  - {ent[0]} ({ent[1]})")
    else:
        print("  - No relevant entities found.")

    # --- Sentiment Analysis using TextBlob ---
    polarity = TextBlob(review).sentiment.polarity
    sentiment = "Positive" if polarity > 0 else "Negative" if polarity < 0 else "Neutral"
    print(f"\nðŸ’¬ Sentiment: {sentiment} ({polarity:.2f})")
    print("-" * 60)
```

```python
# ðŸ›  Task 3: NLP with spaCy + TextBlob

# Required installations (run in Jupyter or CMD before running this script):
# !pip install spacy textblob pandas nltk
# !python -m spacy download en_core_web_sm

import pandas as pd
import spacy
from textblob import TextBlob

# âœ… Load spaCy English model
nlp = spacy.load("en_core_web_sm")

# âœ… Load Amazon Reviews CSV
# Ensure the file name matches your download
df = pd.read_csv("Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products.csv", encoding="ISO-8859-1")

# âœ… Check available columns
print("\nðŸ“„ Available Columns:", df.columns.tolist())

# âœ… Choose the column with review text
review_col = 'reviews.text'  # Change this if the column is different

# âœ… Drop missing reviews and sample 5
sample_reviews = df[review_col].dropna().sample(5, random_state=42).tolist()

# âœ… Process each review
for i, review in enumerate(sample_reviews):
    print(f"\nðŸ“ Review {i+1}:\n{review}\n")

    # --- Named Entity Recognition (NER) ---
    doc = nlp(review)
    product_entities = [(ent.text, ent.label_) for ent in doc.ents if ent.label_ in ['PRODUCT', 'ORG']]

    print("ðŸ§  Named Entities (Product Names / Brands):")
    if product_entities:
        for ent_text, ent_label in product_entities:
            print(f"  - {ent_text} ({ent_label})")
    else:
        print("  - No product/brand entities found.")

    # --- Sentiment Analysis using TextBlob ---
    polarity = TextBlob(review).sentiment.polarity
    sentiment = "Positive" if polarity > 0 else "Negative" if polarity < 0 else "Neutral"

    print(f"\nðŸ’¬ Sentiment: {sentiment} ({polarity:.2f})")
    print("-" * 60)
```
